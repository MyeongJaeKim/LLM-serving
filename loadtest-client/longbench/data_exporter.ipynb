```python
# https://huggingface.co/datasets/THUDM/LongBench/resolve/main/data.zip?download=true

import pandas as pd
import json

lines = []

with open(r'multi_news_e.jsonl') as f:
    lines = f.read().splitlines()

line_dict = [json.loads(line) for line in lines]

df = pd.DataFrame(line_dict)
```

```python
df_cut = df[(df['length'] > 700) & (df['length'] < 1300)].sort_values(by='length').reset_index()
df_cut = df_cut[['context', 'length']]
#df_cut['context'].replace('\n', '', inplace=True)

df_cut['context'] = df_cut['context'].str.replace('NEWLINE_CHAR', ' ')
df_cut['context'] = df_cut['context'].str.replace('\n', ' ')
df_cut
```

```python
df_cut.to_csv('./longbench_multinews_700to1300_len.csv', index=False, header=False)
```
